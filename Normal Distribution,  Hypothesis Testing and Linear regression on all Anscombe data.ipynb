```python
def permutation_sample(var1,var2):
    data=np.concatenate([var1,var2])
    permutation_data=np.random.permutation(data)
    permutate_var1=permutation_data[:len(var1)]
    permutate_var2=permutation_data[len(var2):]
    return permutate_var1,permutate_var2
permutation_sample(rain_june,rain_november)
for i in np.arange(50):
    permutate_var1,permutate_var2=permutation_sample(rain_june,rain_november)
    x,y=ecdf(permutate_var1)
    x1,y1=ecdf(permutate_var2)
    plot(x,y,marker='.',linestyle='none',color='b',alpha=0.03)
    plot(x1,y1,marker='.',linestyle='none',color='r',alpha=0.03)
a,b=ecdf(rain_june)
a1,b1=ecdf(rain_november)
plot(a,b,color='b')
plot(a1,b1,color='r')
```
